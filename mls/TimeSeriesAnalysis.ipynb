```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import torch
```

```python
df = pd.read_csv('../data/csv/main_data.csv')
df = df.drop_duplicates(subset=['date'], keep='first')
df = df.reset_index(drop=True)

df=df[df['date']>='2024-01-01']

df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date').reset_index(drop=True)
```

```python
plt.figure(dpi=240, figsize=(10,7))
plt.plot(df['date'], df['close'], color="#3fc1b0", linewidth=2)
plt.xticks(rotation=90)
plt.grid()
```

```python
df['close_fft'] = np.fft.fft(df['close'])
```

```python
plt.figure(dpi=240)
n = 30
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].plot(df['date'], df['close_fft'], color = '#3fc1b0')
ax[1].plot(df['date'][:n], df['close_fft'][:n], color = '#3fc1b0')
ax[1].set_xticklabels(df['date'][:n], rotation=90)
ax[0].set_xticklabels(df['date'], rotation=90)
ax[1].grid()
ax[0].grid()
```
